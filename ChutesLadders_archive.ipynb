```python
# import statements
import pandas as pd
import random
```

```python
columnNames = ['Turn','DieRoll','PositionStart','PostionEnd']
ChutesLaddersDF = pd.DataFrame(columns=columnNames)
```

```python
def SingleGameChutesAndLadders():
    
    masterList = []
    my_list = [1,0,0,1]
    i = 1

    ladderStart = [1,4,9,21,28,36,51,71,80]
    ladderEnd = [38,14,31,42,84,44,67,91,100]

    chuteStart = [16,47,49,56,62,64,93,95,98]
    chuteEnd = [6,26,11,53,19,60,73,75,78]


    while my_list[3]<100: 
        #turn counter
        my_list[0] = i

        #die roll
        die = random.randint(1,6)
        my_list[1] = die

        #PositionStart Update
        my_list[2] = my_list[3]

        #PostitionEnd Update

        if(die+my_list[3])>100: #Can't exceed 100 logic
            my_list[3]
        else:
            my_list[3] =+ die + my_list[3]

        #print(my_list)


    #For both chutes & Ladders
    #If position end, is on the start of a chute/ladder
    #-take its index
    #and reset PositionEnd to the associated chute/ladder end list

     #Chute List
        if(my_list[3] in chuteStart):
            #print("You Hit a Chute")
            #print("Your Number is:",my_list[3])
            my_list[3]=chuteEnd[chuteStart.index(my_list[3])]
            #print("Your new spot is",my_list[3])   
     #Ladder List
        if(my_list[3] in ladderStart):
            #print("You Hit a Ladder")
            #print("Your Number is:",my_list[3])
            my_list[3]=ladderEnd[ladderStart.index(my_list[3])]
            #print("Your new spot is",my_list[3])

        temp_list = my_list.copy() #needed to copy the list, because was having assignment issues
        masterList.append(temp_list)
        i=i+1
    ChutesLaddersDF = pd.DataFrame(masterList, columns=columnNames)
    return(ChutesLaddersDF)
```

```python
df = SingleGameChutesAndLadders()
```

```python
def MultipleGamesChutesAndLadders(NumGames):
    df_master = SingleGameChutesAndLadders()
    df_master['Game'] = 1
    for i in range(1,NumGames):
        df_temp = SingleGameChutesAndLadders()
        df_temp['Game'] = i + 1
        df_master = df_master.append(df_temp)
    df_master.reset_index(inplace = True)
    df_master.to_csv(str(NumGames)+"Chutes&LaddersGamesSimulated.csv")  
    return(df_master)
```
